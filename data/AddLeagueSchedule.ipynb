# Import

```python
import pandas as pd
```

```python
df = pd.read_csv("./csv/league_schedule.csv",sep=";",encoding="utf-8")
```

```python
df['id']=0
```

```python
df
```

```python
for idx, item in df.iterrows():
    df.loc[idx,'id'] = idx+1
```

```python
def change_team(elem):
    if elem =="T1":
        return 1
    if elem =="GEN":
        return 2
    if elem =="DRX":
        return 3
    if elem =="DK":
        return 4
    if elem =="BRO":
        return 5
    if elem =="KDF":
        return 6
    if elem =="HLE":
        return 7
    if elem =="KT":
        return 8
    if elem =="LSB":
        return 9
    if elem =="NS":
        return 10
    else:
        return 0 
```

```python
df['team1_id']=df['team1_id'].apply(change_team)
df['team1_id']=df['team1_id'].astype('int64')
df['team2_id']=df['team2_id'].apply(change_team)
df['team2_id']=df['team2_id'].astype('int64')
```

```python
df
```

```python
df.to_csv("./csv/league_schedule_final.csv",index=False,sep=";",encoding="utf-8-sig")
```
