```python
import pandas as pd
```

```python
df=pd.read_csv("./csv/player_status.csv",encoding="utf-8",sep=";")
df = df.drop(columns = ["grade"])
```

```python
df['total']=df['lane_status']+df['operation_status']+df['fight_status']+df['ganking_status']+df['roaming_status']+df['jungling_status']
for idx, item in df.iterrows():
        position = item['position']
        df.loc[idx,'position'] = position.upper()
```

## preprocess

```python
def minmax_norm(df_input):
    return (df_input - df_input.min()) / ( df_input.max() - df_input.min())

def normalize(df_input):
    return (df_input - df_input.mean()) / df_input.std()
def give_level(x):
    if x>=1.28:
        return 90
    elif x>=0.84:
        return 80
    elif x>=0.25:
        return 70
    elif x>=-0.53:
        return 60
    else:
        return 50
```

```python
def give_tier(x):
    ## 10퍼
    if x>=1.28:
        return "UR"
    ## 20퍼
    elif x>=0.84:
        return "SR"
    #40퍼
    elif x>=0.25:
        return "R"
    #70퍼
    elif x>=-0.53:
        return "N"
    else:
        return "C"
```

```python
def player_preprocess(df):
    norm_grade=minmax_norm(df['total'])

    norm_grade = normalize(norm_grade)
    norm_grade = pd.DataFrame({'grade':norm_grade})
    norm_grade = norm_grade['grade'].apply(give_tier)
    df = pd.concat([df,norm_grade],axis=1)

    
    

    
    
    

    

    # df = df.rename(columns = {'Player':'nick_name'})
    # df = df.rename(columns = {'Pos':'position'})
    # df = df.rename(columns = {'Team':'base_team_id'})

    return df
```
