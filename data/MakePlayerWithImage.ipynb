# Import

```python
import pandas as pd
```

```python
df = pd.read_csv("./csv/player_no_image.csv",sep=";",encoding="utf-8")
```

```python
df
```

```python
df.dropna()
```

```python
for idx, item in df.iterrows():
    df.loc[idx,"id"]=idx+1
```

```python
df
```

```python
df['id']=df['id'].astype("int64")
df["image"] = "image"
```

```python
url = "https://lck-esports-manager.s3.ap-northeast-2.amazonaws.com/player/"
```

```python
for idx, item in df.iterrows():
    name = item['nick_name']
    df.loc[idx,"image"]=url+name+".png"
```

```python
df
```

```python
df.to_csv("./csv/player.csv",encoding="utf-8-sig",sep=";",index=False)
```
