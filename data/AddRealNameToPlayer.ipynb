```python
import requests
from bs4 import BeautifulSoup
import pandas as pd;
import urllib.request
```

```python
url = "https://lol.inven.co.kr/dataninfo/match/playerTotal.php?category=&category2=169&shipcode=&shipgroup="
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
```

```python
response = requests.get(url,headers=headers);
```

```python
df = pd.read_csv("./csv/player_no_real_name.csv",encoding="utf-8",sep=";")
```

```python
df
```

```python
if response.status_code == 200:
    html = response.text
    soup =BeautifulSoup(html,'html.parser')
    print(soup)
else : 
    print(response.status_code)
```

```python
player_list = soup.find("tbody").find_all("tr")
name_dic = dict()
```

```python
for player in player_list:
    names=player.find_all("td","left")
    real_name = names[0].find("a").text
    nick_name = names[1].find("a").text
    name_dic[nick_name] = real_name
    print(real_name,nick_name)
```

```python
for idx, item in df.iterrows():
    name = item['nick_name']
    if name == "DuDu":
        df.loc[idx,'real_name'] = name_dic["Dudu"]
    elif name == "OnFleek":
        df.loc[idx,'real_name'] = name_dic["Onfleek"]
    elif name =="UmTi":
        df.loc[idx,'real_name'] = name_dic["Umti"]
    elif name == "KiIn":
        df.loc[idx,'real_name'] = name_dic["Kiin"]
    elif name == "FATE":
        df.loc[idx,'real_name'] = name_dic["Fate"]
    elif name == "ShowMaker":
        df.loc[idx,'real_name'] = name_dic["Showmaker"]   
    elif name == "VicLa":
        df.loc[idx,'real_name'] = name_dic["Vicla"]     
    elif name == "deokdam":
        df.loc[idx,'real_name'] = name_dic["Deokdam"]
    elif name == "SamD":
        df.loc[idx,'real_name'] = name_dic["Samd"]
    elif name == "BeryL":
        df.loc[idx,'real_name'] = name_dic["Beryl"]
    elif name == "SnowFlower":
        df.loc[idx,'real_name'] = name_dic["Snowflower"]
    else:
        df.loc[idx,'real_name'] = name_dic[name]
```
