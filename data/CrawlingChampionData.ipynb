# Import

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd;
import urllib.request
```

## Get html

```python
url = "https://lol.inven.co.kr/dataninfo/champion/"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
```

```python
response = requests.get(url,headers=headers);
```

```python
df=pd.DataFrame(columns=["id","korean_name","english_name","position","image","tier","feature"])
```

```python
if response.status_code == 200:
    html = response.text
    soup =BeautifulSoup(html,'html.parser')
    print(soup)
else : 
    print(response.status_code)
```

## Get champion list

```python
champion_list_html = soup.find("ul",id="champListUL")
```

```python
champion_list = champion_list_html.find_all("a")
print(champion_list)
```

## Add data to df

```python
for idx,champion in enumerate(champion_list):
    temp_name = champion['title']
    korean_name = temp_name.split('(')[0].replace(" ","")
    english_name = temp_name.split('(')[1].strip(")").lower().replace("\'","").replace(" ","").replace("\\","").replace('.',"").replace("’","")
    print(english_name)
    test={
        "id":int(idx+1),
        'korean_name':korean_name,
        'english_name':english_name
    }
    df.loc[idx]=test
```

```python
df['id']=df['id'].astype(pd.np.int64)
```

## Get position From fow.kr

```python
url2 = "https://fow.kr/stats"
```

```python
response2 = requests.get(url2,headers=headers);
```

```python
if response2.status_code == 200:
    html = response2.text
    soup =BeautifulSoup(html,'html.parser')
    print(soup)
else : 
    print(response2.status_code)
```

```python
champion_table = soup.find("table")
```

```python
champion_list = champion_table.find_all("tr",class_="champ_one")
```

```python
champion_list
```

```python
for champion in champion_list:
    position = champion['position']
    if position =="utility":
        position = "support"
    if position == "bottom":
        position = "adc"
    position = position.upper()
    korean_name = champion['rname']
    print(korean_name,position)
    df.loc[df["korean_name"]== korean_name,"position"] = position
```

## Get Image from op.gg

```python
url3 = "https://www.op.gg/champions/"
```

```python
response3 = requests.get(url3,headers=headers);
```

```python
if response3.status_code == 200:
    html = response3.text
    soup =BeautifulSoup(html,'html.parser')
```

```python
champion_list_html = soup.find("nav","css-1x3kezq")
```

```python
champion_list = champion_list_html.find_all("img")
```

```python
for champion in champion_list:
    english_name = champion['alt'].replace("\'","").replace(" ","").replace("'\'","").replace('.',"")
    if english_name == "monkeyking":
        english_name ="wukong"
    if english_name == "renata":
        english_name ="renataglasc"
    image = champion['src'].split('?')[0]
    df.loc[df['english_name'] == english_name,'image']=image
```

```python
df.to_csv("champion_no_tier.csv", index=False, encoding="utf-8-sig",sep=";")
```
